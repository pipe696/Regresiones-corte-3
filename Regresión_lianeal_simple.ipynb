```python
# prompt: necesito un análisis de regresión lineal simple que la variable X sea "Salario" y  la variable Y sea "Tasa _Crecimiento" de nuestra base de datos  "datos"

import pandas as pd
import statsmodels.api as sm

# Cargar los datos (asumiendo que ya tienes el archivo en tu Google Drive)
datos = pd.read_excel("/content/drive/MyDrive/estadística /corte 3 /datos_industrias.xlsx")

# Preparar los datos para la regresión
X = datos["Empleados"]
y = datos["Tasa_Crecimiento"]

# Agregar una constante a la variable independiente (intercepto)
X = sm.add_constant(X)

# Ajustar el modelo de regresión lineal
modelo = sm.OLS(y, X).fit()

# Imprimir los resultados del modelo
print(modelo.summary())
```
